# Lendo base de dados

```python
import pandas as pd

df = pd.read_csv('../dataset/ml-100k/u.item', sep='|', header=None)
df.head()
```

## Criando metadados de itens

```python
df_gen = df.iloc[:,4:]

metadata = []
for row in df_gen.iterrows():
    for g, gen in enumerate(row[1]):
        item = df[0][row[0]]
        if gen != 0:
            metadata.append((item, g))
    

gen_df = pd.DataFrame(metadata)

gen_df.to_csv('../dataset/ml-100k/item_metadata.dat', header=None, sep='\t', index=False)
```

## Lendo arquivo de interações e substituindo id por nome do filme

```python
df_rating = pd.read_csv('../dataset/ml-100k/u.data', sep='\t', header=None)
df_rating[1] = df_rating[1].map(df.set_index(0)[1])
df_rating[df_rating[0] == 1]
```
